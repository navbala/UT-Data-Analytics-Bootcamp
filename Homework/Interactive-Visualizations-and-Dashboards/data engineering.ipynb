```python
# Create engine using the `sqlite` database file created provided in the data set
engine = create_engine("sqlite:///belly_button_biodiversity.sqlite")
```

```python
Base = automap_base()
```

```python
Base.prepare(engine, reflect=True)
Base.classes.keys()
```
